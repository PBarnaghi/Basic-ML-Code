A simple demostration to show how changing the coorindate system can help to separate the data

```matlab
Tbl = readtable('blob-data.csv');
                                  
data = table2array(Tbl);
```

```matlab
scatter(data(:,1), data(:,2), 20, data(:,3)); axis square;
```

```matlab
data(1:3, 1:end)
```

```matlab
Tbl(1:3, 1:end)
```

```matlab
polar_data_raduis = sqrt(data(1:end, 1).^2 + data(1:end, 2).^2);
polar_data_theta = atan(data(1:end, 2:2)./data(1:end, 1));
%Converting theta from radian to degree
polar_data_theta = 180 * polar_data_theta/pi;
```

```matlab
scatter(polar_data_raduis, polar_data_theta, 20, data(:,3)); axis square;
```
